```python
import pandas as pd

# Load the dataset
df = pd.read_csv("house_dataset.csv")

# Check Missing Values
print("Missing Values in Each Column:")
missing_values = df.isnull().sum()
print(missing_values)

# Calculate percentage of missing values
missing_percentage = (missing_values / len(df)) * 100

print("\nPercentage of Missing Values:")
print(missing_percentage)

# Handle Missing Values
# Fill missing values with column mean (for numeric columns)

df = df.fillna(df.mean(numeric_only=True))

print("\nMissing Values After Filling:")
print(df.isnull().sum())

# Check Duplicate Rows
duplicates = df.duplicated().sum()

print("\nNumber of Duplicate Rows:", duplicates)

# Remove duplicate rows
df = df.drop_duplicates()

print("Shape After Removing Duplicates:", df.shape)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("house_dataset.csv")

# List of important numeric columns
important_columns = [
    "Feature1",
    "Feature2",
    "Feature3",
    "Feature4",
    "Feature5",
    "HousePrice"
]

# Plot histograms
for col in important_columns:
    plt.figure(figsize=(6,4))
    plt.hist(df[col], bins=10, edgecolor='black')
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.grid(alpha=0.3)
    plt.show()
```

```python
import seaborn as sns

# Create Categorical Features
# House Size Category
df['HouseSize'] = pd.cut(df['Feature1'],
                         bins=[0, 33, 66, 100],
                         labels=['Small', 'Medium', 'Large'])

# Location Category
df['Location'] = pd.cut(df['Feature2'],
                        bins=[0, 33, 66, 100],
                        labels=['Urban', 'Suburban', 'Rural'])

# Construction Quality
df['Quality'] = pd.cut(df['Feature3'],
                       bins=[0, 33, 66, 100],
                       labels=['Low', 'Medium', 'High'])

# Identify Categorical Columns
categorical_columns = df.select_dtypes(include=['object', 'category']).columns
print("Categorical Columns:")
print(categorical_columns)

# Count Plots
for col in categorical_columns:
    plt.figure(figsize=(6,4))
    sns.countplot(data=df, x=col)
    plt.title(f"Count Plot of {col}")
    plt.xticks(rotation=20)
    plt.show()
```

```python

import seaborn as sns

# Correlation Matrix
corr_matrix = df.corr(numeric_only=True)

# Correlation Heatmap
plt.figure(figsize=(20, 16))
sns.heatmap(corr_matrix,
            cmap='coolwarm',
            annot=False,
            linewidths=0.5)

plt.title("Correlation Heatmap of Numeric Features")
plt.show()

# Correlation with Target Variable
target = "HousePrice"

target_corr = corr_matrix[target].sort_values(ascending=False)

print("\nCorrelation of Features with HousePrice:\n")
print(target_corr)
```

```python

# Create Sample Categorical Columns

df['HouseSize'] = pd.cut(df['Feature1'],
                         bins=[0,33,66,100],
                         labels=['Small','Medium','Large'])

df['Location'] = pd.cut(df['Feature2'],
                        bins=[0,33,66,100],
                        labels=['Urban','Suburban','Rural'])

# Before Encoding

print("Before Encoding:")
print(df[['HouseSize','Location']].head())

# One-Hot Encoding

df_encoded = pd.get_dummies(df,
                            columns=['HouseSize','Location'],
                            dtype=int)

# After Encoding

print("\nAfter Encoding:")
print(df_encoded.head())

print("\nShape Before Encoding:", df.shape)
print("Shape After Encoding:", df_encoded.shape)
```

```python
from sklearn.preprocessing import StandardScaler

# Independent Features (X)
# Use df_encoded which has numerical representations of categorical features
# Drop 'HousePrice' (target) and 'Quality' (original categorical column still present)
X = df_encoded.drop(["HousePrice", "Quality"], axis=1)

# Apply Standard Scaling
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

# Convert scaled data into DataFrame
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

# Display first 5 rows
print("First 5 Rows of Scaled Features:")
print(X_scaled.head())
```
